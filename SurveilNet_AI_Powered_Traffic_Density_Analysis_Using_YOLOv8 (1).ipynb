```python
import os
import zipfile
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ultralytics import YOLO
from google.colab.patches import cv2_imshow

# Path to dataset ZIP file
zip_file_path = "/content/New Compressed (zipped) Folder.zip"

# Destination folder for extracted images
extract_folder = "/content/extracted_traffic_images"
os.makedirs(extract_folder, exist_ok=True)

# Extract ZIP file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_folder)

print(f"Dataset extracted to: {extract_folder}")

# Load YOLOv8 model
model = YOLO("yolov8n.pt")

# Get image files from extracted dataset (limit to 15)
image_files = [f for f in os.listdir(extract_folder) if f.endswith('.jpg')][:15]

# Store density values for visualization
density_values = []

if not image_files:
    print("No image files found in dataset.")
else:
    for image_name in image_files:
        image_path = os.path.join(extract_folder, image_name)
        image = cv2.imread(image_path)

        # Object detection using YOLOv8
        results = model(image)

        # Count detected objects
        num_objects = len(results[0].boxes)
        density_values.append(num_objects)

        # Render detections
        annotated_image = results[0].plot()

        # Display processed image (Google Colab Compatible)
        cv2_imshow(annotated_image)
        cv2.waitKey(500)  # Show each image for 500ms

cv2.destroyAllWindows()
print("Processing complete for first 15 images!")

# Plot traffic density trend
plt.plot(density_values, label="Traffic Density")
plt.xlabel("Image Number")
plt.ylabel("Number of Vehicles/Pedestrians")
plt.title("Traffic Density Analysis for First 15 Images")
plt.legend()
plt.show()
```
